```python
import pandas as pd
import numpy as np
```

```python
prod = pd.read_csv('data/GlobalFoodProduction.csv')
prod.head()
```

```python
print(prod.columns)
```

```python
# print unique values in the item column
print(prod['Item'].unique())
```

```python
# drop the rows with item values of "Sweeteners, Other" and "Sugar & Sweeteners" and "Sesame seed" and 'Coffee and products' 'Cocoa Beans and products' 'Tea (including mate)' 'Pepper' 'Spices, Other' 'Wine' 'Beer' 'Beverages, Alcoholic' 'Oilcrops' 'Infant food' 'Beverages, Fermented'
prod = prod[~prod['Item'].isin(['Sweeteners, Other', 'Sugar & Sweeteners', 'Sesame seed', 'Coffee and products', 'Cocoa Beans and products', 'Tea (including mate)', 'Pepper', 'Spices, Other', 'Wine', 'Beer', 'Beverages, Alcoholic', 'Oilcrops', 'Infant food', 'Beverages, Fermented'])]
```

```python
# print unique values in the item column
print(prod['Item'].unique())
```

```python
# drop the rows with item values of 'Stimulants' 'Spices' 'Alcoholic Beverages' "Oilcrops, Other" 'Rape and Mustardseed' 'Cottonseed' 'Sugar non-centrifugal' 'Palm kernels' 'Cloves' 'Sunflower seed'
prod = prod[~prod['Item'].isin(['Stimulants', 'Spices', 'Alcoholic Beverages', "Oilcrops, Other", 'Rape and Mustardseed', 'Cottonseed', 'Sugar non-centrifugal', 'Palm kernels', 'Cloves', 'Sunflower seed'])]
```

```python
# print unique values in the item column
print(prod['Item'].unique())
```

```python
prod.drop([x for x in prod.columns if x.__contains__("Y") and x != "Y2013"], axis=1, inplace=True)
prod
```

```python
# if the value in the column "Item" contains "Oil", then set the value in the column "Item" to "Oil"
prod.loc[prod['Item'].str.contains("Oil"), 'Item'] = "Oil"
```

```python
dict1 = { col : "first" for col in prod.columns if col != "Y2013"}
dict1['Y2013'] = "sum"
dict1
```

```python
# drop all rows where element == feed
prod = prod[prod['Element'] != 'Feed']
prod.drop(["Element Code", "Element"], axis=1, inplace=True)
```
