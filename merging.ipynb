# Cleaning data for efficient mapping

```python
import pandas as pd
```

```python
pop = pd.read_csv('./data/Pop.csv')
prod = pd.read_csv('./data/GlobalFoodProduction.csv')
cal = pd.read_csv('./data/Cal.csv')
```

```python
cal["x1"] = cal["x1"].apply(lambda x : x.lower())
pop["x1"] = pop["x1"].apply(lambda x : x.lower())
prod["Area"] = prod["Area"].apply(lambda x : x.lower())
```

```python
# find rows of prod where Area contains china
prod[prod["Area"].str.contains("china")].Area.unique()
# change china to china, mainland
prod["Area"] = prod["Area"].apply(lambda x : x.replace("china, mainland", "china"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("china, hong kong sar", "hong kong"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("china, taiwan province of", "taiwan"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("china, macao sar", "macao"))
```

```python
prod["Area"] = prod["Area"].apply(lambda x : x.replace("united states of america", "united states"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("russian federation", "russia"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("iran (islamic republic of)", "iran"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("lao people's democratic republic", "laos"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("republic of moldova", "moldova"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("venezuela (bolivarian republic of)", "venezuela"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("viet nam", "vietnam"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("brunei darussalam", "brunei"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("bolivia (plurinational state of)", "bolivia"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("united republic of tanzania", "tanzania"))
prod["Area"] = prod["Area"].apply(lambda x : x.replace("the former yugoslav republic of macedonia", "north macedonia"))
```

```python
# find rows of prod where area contains
prod[prod["Area"].str.contains("world")].Area.unique()
```

```python
cal["x1"] = cal["x1"].apply(lambda x : x.replace("united states of america", "united states"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("russian federation", "russia"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("iran (islamic republic of)", "iran"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("lao people's democratic republic", "laos"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("republic of moldova", "moldova"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("venezuela (bolivarian republic of)", "venezuela"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("viet nam", "vietnam"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("brunei darussalam", "brunei"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("bolivia (plurinational state of)", "bolivia"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("united republic of tanzania", "tanzania"))
cal["x1"] = cal["x1"].apply(lambda x : x.replace("the former yugoslav republic of macedonia", "north macedonia"))
```

```python
# add a row to pop
pop = pop.append({"x1": "bahamas", "x2": 328289}, ignore_index=True)
pop = pop.append({"x1": "cape verde", "x2": 530549}, ignore_index=True)
pop = pop.append({"x1": "congo", "x2": 4574052}, ignore_index=True)
pop = pop.append({"x1": "cote d'ivoire", "x2": 23284095}, ignore_index=True)
pop = pop.append({"x1": "democratic republic of congo", "x2": 80747292}, ignore_index=True)
pop = pop.append({"x1": "gambia", "x2": 1913150}, ignore_index=True)
pop = pop.append({"x1": "macao", "x2": 586116}, ignore_index=True)
pop = pop.append({"x1": "micronesia", "x2": 105995}, ignore_index=True)
pop = pop.append({"x1": "myanmar", "x2": 53044168}, ignore_index=True)
pop = pop.append({"x1": "north korea", "x2": 24767164}, ignore_index=True)
pop = pop.append({"x1": "south korea", "x2": 50063069}, ignore_index=True)
pop = pop.append({"x1": "timor", "x2": 1172274}, ignore_index=True)
```

```python
tot = cal.merge(pop, left_on="x1", right_on="x1", how="inner")
```

```python
tot = tot.merge(prod, left_on="x1", right_on="Area", how="inner")
```

```python
tot
#rename columns in tot from x1 to country, x2_x to population, x2_y to calories, and x2 to production
tot = tot.rename(columns={"x2_x": "fats", "x3": "proteins", "x4": "carbs", "x5": "fruits_and_veggies", "x2_y": "population"})
tot.drop("x1", axis=1, inplace=True)
```
